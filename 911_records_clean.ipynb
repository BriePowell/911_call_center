```python
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
```

```python
data.tail()
```

```python
data.set_index('id', inplace=True)
data.head()
```

```python
seriesAlpha = data['description']
```

```python
alpha = seriesAlpha.to_list()
```

```python
beta = [a for a in alpha if str(a) != 'nan']
```

```python
beta
```

```python
gamma = [sub.split() for sub in beta]
```

```python
print(gamma) #word tokens
```

```python
# Update stop words

['county', 'dispatcher', 'woman', 'year-old', '2009', '2008','(fla.)', 'report', 'later', 'call', '2010', '2011','2012', '2013']
```

```python
stp_wrds = stopwords.words('english')

go_wrds = []
for li in gamma:
    for wrd in li:
        if wrd.lower() not in stp_wrds:
            go_wrds.append(wrd.lower())

go_wrds

```

```python
len(go_wrds)
```

```python
tokens = pd.DataFrame({'word token': go_wrds})

print(tokens)
```

```python
top_words = tokens.groupby(['word token']).size().reset_index(name='counts')
top_words = top_words.sort_values(by = 'counts', ascending = False)
top_words
```
